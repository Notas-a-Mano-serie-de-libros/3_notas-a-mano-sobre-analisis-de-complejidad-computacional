```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.optimize import brentq

plt.style.use('default')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['savefig.edgecolor'] = 'white'

plt.rcParams['figure.dpi'] = 500
plt.rcParams['savefig.dpi'] = 500

plt.rcParams['savefig.pad_inches'] = 0.05

plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
```

```python
# Definir las funciones C(n) y g(n)
def C(n):
    return n**3 + 2 * n**2 + n + 5

def g(n):
    return n**4

# Constante c
c = 10

# Definir función diferencia para hallar el punto de cruce
def diff(n):
    return c * g(n) - C(n)

# Calcular n₀ de forma precisa usando búsqueda de raíces
n0 = brentq(diff, 0.1, 10)  # buscamos entre 0.1 y 10 para evitar valores triviales
n0_ceil = math.ceil(n0)

# Rango amplio para visualizar cómo la cota domina para n grandes
n = np.linspace(0, 1e20, 1000000)

# Calcular los valores de las funciones
f_values = C(n)
g_values = c * g(n)

# Crear la gráfica
fig, ax = plt.subplots(figsize=(8, 3))

# Etiqueta y título
label_text = f'$c \\cdot g(n) = {c}n^4$'
title_text = '$C(n)$ vs $o(g(n))$'

# Graficar las funciones
ax.plot(n, f_values, label='$C(n) = n^3 + 2n^2 + n + 5$', color='blue')
ax.plot(n, g_values, label=label_text, color='red')

# Línea vertical en n₀
ax.axvline(x=n0, color='green', linestyle='--', label=rf'$n_0 \approx \lceil {n0:.2f} \rceil = {n0_ceil}$')

# Área donde se cumple la desigualdad
ax.fill_between(n, f_values, g_values, where=(f_values <= g_values), interpolate=True, color='red', alpha=0.3)

# Etiquetas y título
ax.set_xlabel('Tamaño de la entrada ($n$)')
ax.set_ylabel('$o(g(n))$')
ax.set_title(title_text)
ax.legend(loc='upper left', bbox_to_anchor=(0.06, 1))
ax.grid(True)

# Mostrar y guardar la gráfica
plt.show()
fig.savefig('graficas/comparacion_little_o.png', bbox_inches='tight')
```

```python
# Definir las funciones C(n) y g(n)
def C(n):
    return n**3 + 2 * n**2 + n + 5

def g(n):
    return n**4

# Constante c
c = 10

# Definir función diferencia para hallar el punto de cruce
def diff(n):
    return c * g(n) - C(n)

# Calcular n₀ de forma precisa usando búsqueda de raíces
n0 = brentq(diff, 0.1, 10)  # buscamos entre 0.1 y 10
n0_ceil = math.ceil(n0)

# Rango amplio en escala logarítmica (evitamos 0 porque log(0) no está definido)
n = np.logspace(-1, 20, 1000)  # desde 10^-1 hasta 10^20

# Calcular los valores de las funciones
f_values = C(n)
g_values = c * g(n)

# Crear la gráfica
fig, ax = plt.subplots(figsize=(8, 3))

# Escala logarítmica
ax.set_xscale('log')
ax.set_yscale('log')

# Etiqueta y título
label_text = f'$c \\cdot g(n) = {c}n^4$'
title_text = '$C(n)$ vs $o(g(n))$'

# Graficar las funciones
ax.plot(n, f_values, label='$C(n) = n^3 + 2n^2 + n + 5$', color='blue')
ax.plot(n, g_values, label=label_text, color='red')

# Línea vertical en n₀
ax.axvline(x=n0, color='green', linestyle='--', label=rf'$n_0 \approx \lceil {n0:.2f} \rceil = {n0_ceil}$')

# Área donde se cumple la desigualdad
ax.fill_between(n, f_values, g_values, where=(f_values <= g_values), interpolate=True, color='red', alpha=0.3)

# Etiquetas y título
ax.set_xlabel('Tamaño de la entrada ($n$)')
ax.set_ylabel('$o(g(n))$')
ax.set_title(title_text)
ax.legend(loc='upper left', bbox_to_anchor=(0.1, 1))
ax.grid(True, which='both', linestyle='--', linewidth=0.5)

# Mostrar y guardar la gráfica
plt.show()
fig.savefig('graficas/comparacion_little_o2.png', bbox_inches='tight')
```
