```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.optimize import brentq

plt.style.use('default')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['savefig.edgecolor'] = 'white'

plt.rcParams['figure.dpi'] = 500
plt.rcParams['savefig.dpi'] = 500

plt.rcParams['savefig.pad_inches'] = 0.05

plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
```

```python
# Definir las funciones
def C(n):
    return n**3 + 2*n**2 + n + 5

def g(n):
    return n**3

# Constante c para Omega
c = 0.5

# Función diferencia
def diff(n):
    return C(n) - c * g(n)

# Evaluar extremos
f1 = diff(1)
f2 = diff(1e6)

# Determinar si hay cruce de signo
try:
    if f1 * f2 < 0:
        n0 = brentq(diff, 1, 1e6)
    else:
        n0 = 0
except ValueError:
    n0 = 0

n0_ceil = math.ceil(n0)

# Crear valores de n
n1 = np.linspace(1, 1e4, 5000)
n2 = np.linspace(1e4, 1e20, 5000)
n = np.concatenate((n1, n2))

# Evaluar funciones
f_values = C(n)
g_values = c * g(n)

# Crear la gráfica
fig, ax = plt.subplots(figsize=(8, 3))

# Etiquetas
label_text = f'$c \\cdot g(n) = {c}n^3$'
title_text = '$C(n)$ vs $\\Omega(g(n))$'

# Graficar
ax.plot(n, f_values, label='$C(n) = n^3 + 2n^2 + n + 5$', color='blue')
ax.plot(n, g_values, label=label_text, color='orange')
ax.axvline(x=n0, color='green', linestyle='--', label=f'$n_0 = {n0_ceil}$')
ax.fill_between(n, g_values, f_values, where=(f_values >= g_values), color='blue', alpha=0.3)

# Etiquetas
ax.set_xlabel('Tamaño de la entrada ($n$)')
ax.set_ylabel('$\\Omega(g(n))$')
ax.set_title(title_text)
ax.legend(loc='upper left', bbox_to_anchor=(0.06, 1))
ax.grid(True)

# Mostrar y guardar
plt.show()
fig.savefig('graficas/comparacion_big_omega.png', bbox_inches='tight')
```

```python
def C(n):
    return n**3 + 2*n**2 + n + 5

def g(n):
    return n**3

# Constante c para Omega
c = 0.5

# Función diferencia
def diff(n):
    return C(n) - c * g(n)

# Evaluar extremos
f1 = diff(1)
f2 = diff(1e6)

# Determinar si hay cruce de signo
try:
    if f1 * f2 < 0:
        n0 = brentq(diff, 1, 1e6)
    else:
        n0 = 0
except ValueError:
    n0 = 0

n0_ceil = math.ceil(n0)

# Crear valores de n en escala logarítmica
n = np.logspace(0, 5, 5000)  # De 10^0 a 10^8

# Evaluar funciones
f_values = C(n)
g_values = c * g(n)

# Crear la gráfica
fig, ax = plt.subplots(figsize=(8, 3))

# Etiquetas
label_text = f'$c \\cdot g(n) = {c}n^3$'
title_text = '$C(n)$ vs $\\Omega(g(n))$'

# Graficar
ax.plot(n, f_values, label='$C(n) = n^3 + 2n^2 + n + 5$', color='blue')
ax.plot(n, g_values, label=label_text, color='orange')
ax.axvline(x=n0, color='green', linestyle='--', label=f'$n_0 = {n0_ceil}$')
ax.fill_between(n, g_values, f_values, where=(f_values >= g_values), color='blue', alpha=0.3)

# Escalas logarítmicas
ax.set_xscale('log')
ax.set_yscale('log')

# Etiquetas y título
ax.set_xlabel('Tamaño de la entrada ($n$)')
ax.set_ylabel('$\\Omega(g(n))$')
ax.set_title(title_text)
ax.legend(loc='upper left', bbox_to_anchor=(0.06, 1))
ax.grid(True)

# Mostrar y guardar
plt.show()
fig.savefig('graficas/comparacion_big_omega_2.png', bbox_inches='tight')
```
