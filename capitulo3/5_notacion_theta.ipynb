## Simulación interactiva del límite asintótico para notación $\Theta$

La notación $\Theta$ define la familia de funciones que están acotadas simultáneamente por abajo y por arriba mediante múltiplos constantes de una misma función de referencia. Su propósito es establecer una **cota ajustada** sobre el crecimiento de $C(n)$, ignorando constantes y términos de menor orden.

Formalmente, una función $C(n)$ pertenece a $\Theta(g(n))$ si existen constantes positivas $c_1$ y $c_2$ tales que:

$$
c_1\cdot g(n)\le C(n)\le c_2\cdot g(n), \qquad \forall n\ge n_0
$$

Esta definición combina simultáneamente las ideas de cota inferior y cota superior:

$$
C(n)\in\Theta(g(n))\Longleftrightarrow C(n)\in\Omega(g(n))\land C(n)\in O(g(n))
$$

Si el cociente entre $C(n)$ y $g(n)$ converge a una constante positiva $k$, entonces las constantes deben elegirse de forma que $C(n)$ quede atrapada entre ambas cotas:

$$
0\lt c_1\le k\le c_2
$$

El valor de $n_0$ se obtiene identificando el primer entero $A$ desde el cual se cumplen simultáneamente ambas desigualdades:

$$
n_0=\min\left\{\lceil A\rceil\in\mathbb{N}\mid \forall n\ge A,\; c_1\cdot g(n)\le C(n)\le c_2\cdot g(n)\right\}
$$

La simulación permite modificar $C(n)$, $g(n)$, $c_1$, $c_2$, los términos de menor orden, el intervalo visible y la escala para observar cuándo una función queda realmente acotada por ambos lados.

## Ejemplo del libro


```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.optimize import brentq

plt.style.use('default')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['savefig.edgecolor'] = 'white'

plt.rcParams['figure.dpi'] = 500
plt.rcParams['savefig.dpi'] = 500

plt.rcParams['savefig.pad_inches'] = 0.05

plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
```

```python
# Definir funciones
def C(n):
    return n**3 + 2*n**2 + n + 5

def g(n):
    return n**3

# Constantes
c1 = 0.5
c2 = 2

# Funciones diferencia para encontrar n0
def diff_lower(n):
    return C(n) - c1 * g(n)

def diff_upper(n):
    return c2 * g(n) - C(n)

# Encontrar los n0 correspondientes
try:
    n0_lower = brentq(diff_lower, 1, 1e6)
except ValueError:
    n0_lower = 0
try:
    n0_upper = brentq(diff_upper, 1, 1e6)
except ValueError:
    n0_upper = 0

# Tomamos el mayor de ambos como verdadero n0 para Big-Theta
n0_theta = max(n0_lower, n0_upper)
n0_ceil = math.ceil(n0_theta)

# Rango de n combinado
n1 = np.linspace(1, 1e4, 5000)
n2 = np.linspace(1e4, 1e20, 5000)
n = np.concatenate((n1, n2))

# Evaluar funciones
f_values = C(n)
g_lower = c1 * g(n)
g_upper = c2 * g(n)

# Crear gráfica
fig, ax = plt.subplots(figsize=(8, 3))

# Etiquetas
title_text = '$C(n)$ vs $\\Theta(g(n))$'
label_c1 = f'$c_1 \\cdot g(n) = {c1}n^3$'
label_c2 = f'$c_2 \\cdot g(n) = {c2}n^3$'

# Graficar funciones
ax.plot(n, f_values, label='$C(n) = n^3 + 2n^2 + n + 5$', color='#1565C0')
ax.plot(n, g_lower, label=label_c1, color='#EF6C00')
ax.plot(n, g_upper, label=label_c2, color='#B71C1C')

# Línea vertical en n0
ax.axvline(x=n0_theta, color='#2E7D32', linestyle='--', label=rf'$n_0 \approx \lceil {n0_theta:.2f} \rceil = {n0_ceil}$')

# Rellenar región válida para Theta
ax.fill_between(n, g_lower, g_upper, where=(f_values >= g_lower) & (f_values <= g_upper), color='#1565C0', alpha=0.25)

# Etiquetas
ax.set_xlabel('Tamaño de la entrada ($n$)')
ax.set_ylabel('$\\Theta(g(n))$')
ax.set_title(title_text)
ax.legend(loc='upper left', bbox_to_anchor=(0.06, 1))
ax.grid(True)

# Mostrar y guardar
plt.show()
fig.savefig('graficas/comparacion_big_theta.png', bbox_inches='tight')
```
