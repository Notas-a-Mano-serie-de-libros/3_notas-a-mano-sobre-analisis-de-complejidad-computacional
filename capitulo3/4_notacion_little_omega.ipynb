## Simulación interactiva del límite asintótico para notación $\omega$

La notación $\omega$ se utiliza para mostrar que $C(n)$ crece **estrictamente más rápido** que una función de referencia $g(n)$. Es la contraparte estricta de la notación $\Omega$: no basta con que $C(n)$ esté por encima de algún múltiplo de $g(n)$, sino que el cociente debe crecer sin límite.

Formalmente, se exige que el cociente diverja hacia infinito:

$$
\lim_{n\to\infty}\left(\frac{C(n)}{g(n)}\right)=\infty
$$

Esto significa que, para cualquier constante positiva $c$, existe un punto $n_0$ a partir del cual se cumple:

$$
\left|\frac{C(n)}{g(n)}\right|\gt c, \qquad \forall n\ge n_0
$$

Despejando $C(n)$, la condición se expresa como:

$$
C(n)\gt c\cdot g(n), \qquad \forall n\ge n_0
$$

El valor de $n_0$ se calcula identificando el primer entero desde el cual la desigualdad estricta se mantiene para todos los valores posteriores:

$$
n_0=\min\left\{\lceil A\rceil\in\mathbb{N}\mid \forall n\ge A,\; C(n)\gt c\cdot g(n)\right\}
$$

La simulación permite observar cómo la función de interés termina superando cualquier múltiplo constante de la función de referencia cuando realmente pertenece a $\omega(g(n))$.

## Ejemplo del libro


```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('default')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['savefig.edgecolor'] = 'white'

plt.rcParams['figure.dpi'] = 500
plt.rcParams['savefig.dpi'] = 500

plt.rcParams['savefig.pad_inches'] = 0.05

plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
```

```python
from scipy.optimize import brentq
import math

# Definir las funciones
def C(n):
    return n**3 + 2*n**2 + n + 5

def g(n):
    return n**2

# Constante c para little omega
c = 1

# Función diferencia
def diff(n):
    return C(n) - c * g(n)

# Evaluar extremos
f1 = diff(1)
f2 = diff(1e6)

# Determinar si hay cruce de signo
try:
    if f1 * f2 < 0:
        n0 = brentq(diff, 1, 1e6)
    else:
        n0 = 0
except ValueError:
    n0 = 0

n0_ceil = math.ceil(n0)

# Crear valores de n
n1 = np.linspace(1, 1e4, 5000)
n2 = np.linspace(1e4, 1e20, 5000)
n = np.concatenate((n1, n2))

# Evaluar funciones
f_values = C(n)
g_values = c * g(n)

# Crear la gráfica
fig, ax = plt.subplots(figsize=(8, 3))

# Etiquetas
label_text = "$c \\cdot g(n) = " + (f"{c}\\cdot " if c != 1 else "") + "n^2$"
title_text = '$C(n)$ vs $\\omega(g(n))$'

# Graficar
ax.plot(n, f_values, label='$C(n) = n^3 + 2n^2 + n + 5$', color='#1565C0')
ax.plot(n, g_values, label=label_text, color='#EF6C00')
ax.axvline(x=n0, color='#2E7D32', linestyle='--', label=f'$n_0 = {n0_ceil}$')
ax.fill_between(n, g_values, f_values, where=(f_values > g_values), color='#1565C0', alpha=0.25)

# Etiquetas
ax.set_xlabel('Tamaño de la entrada ($n$)')
ax.set_ylabel('$\\omega(g(n))$')
ax.set_title(title_text)
ax.legend(loc='upper left', bbox_to_anchor=(0.06, 1))
ax.grid(True)

# Mostrar y guardar
plt.show()
fig.savefig('graficas/comparacion_little_omega.png', bbox_inches='tight')
```
