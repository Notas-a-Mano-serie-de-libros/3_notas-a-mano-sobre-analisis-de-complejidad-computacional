## 1 notacion big o


## Ejemplo del libro


```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from scipy.optimize import brentq

if (Path.cwd() / 'capitulo3').is_dir():
    GRAPH_DIR = Path.cwd() / 'capitulo3' / 'graficas'
elif Path.cwd().name == 'referencias':
    GRAPH_DIR = Path.cwd().parent / 'graficas'
else:
    GRAPH_DIR = Path.cwd() / 'graficas'
GRAPH_DIR.mkdir(parents=True, exist_ok=True)

def format_n0(value):
    nearest_integer = round(float(value))
    if np.isclose(value, nearest_integer):
        return str(nearest_integer)
    return f'{value:.2f}'

plt.style.use('default')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['savefig.edgecolor'] = 'white'

plt.rcParams['figure.dpi'] = 500
plt.rcParams['savefig.dpi'] = 500

plt.rcParams['savefig.pad_inches'] = 0.05

plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
```

```python
# Definir las funciones
def C(n):
    return n**3 + 2*n**2 + n + 5

def g(n):
    return n**3

# Constante c
c = 2

# Definir la función diferencia para buscar n₀
def diff(n):
    return c * g(n) - C(n)

# Calcular n₀ de forma precisa
n0 = brentq(diff, 1, 1e6)  # límite ajustado para evitar overflows

n1 = np.linspace(1, 1e4, 5000)
n2 = np.linspace(1e4, 1e20, 5000)
n = np.concatenate((n1, n2))

# Evaluar funciones
f_values = C(n)
g_values = c * g(n)

# Crear la gráfica
fig, ax = plt.subplots(figsize=(8, 3))

# Títulos y etiquetas
label_text = f'$c \\cdot g(n) = {c}n^3$'
title_text = '$C(n)$ vs $O(g(n))$'

# Graficar funciones
ax.plot(n, f_values, label='$C(n) = n^3 + 2n^2 + n + 5$', color='#1565C0')
ax.plot(n, g_values, label=label_text, color='#B71C1C')

# Línea vertical en n₀
ax.axvline(x=n0, color='#2E7D32', linestyle='--', label=rf'$n_0 = {format_n0(n0)}$')

# Área donde se cumple la cota
ax.fill_between(n, f_values, g_values, where=(f_values <= g_values), color='#B71C1C', alpha=0.3)

# Escala lineal
ax.set_xscale('linear')
ax.set_yscale('linear')

# Etiquetas y leyenda
ax.set_xlabel('Tamaño de la entrada ($n$)')
ax.set_ylabel('$O(g(n))$')
ax.set_title(title_text)
ax.legend(loc='upper left', bbox_to_anchor=(0.06, 1))
ax.grid(True)

# Mostrar y guardar
plt.show()
fig.savefig(GRAPH_DIR / 'comparacion_big_O.png', bbox_inches='tight')
```

```python
# Definir las funciones C(n) y g(n)
def C(n):
    return n**3 + 2*n**2 + n + 5

def g(n):
    return n**3

# Constante c
c = 2

# Rango de valores de n (ajustado para escala logarítmica)
n = np.logspace(0, 20, 1000000)  # de 10^0 a 10^2 => de 1 a 100

# Calcular los valores de las funciones
f_values = C(n)
g_values = c * g(n)

# Encontrar el primer valor de n tal que C(n) <= c * g(n)
mask = f_values <= g_values
if np.any(mask):
    n0_index = np.argmax(mask)
    n0 = n[n0_index]
else:
    n0 = None
    print("No se encontró un valor de n₀ donde C(n) <= c·g(n) en el rango dado.")

# Crear la gráfica
fig, ax = plt.subplots(figsize=(8, 3))

# Configurar la etiqueta de g(n)
if c == 1:
    label_text = '$c \\cdot g(n) = n^3$'
    title_text = '$C(n)$ vs $O(g(n))$'
else:
    label_text = f'$c \\cdot g(n) = {c}n^3$'
    title_text = '$C(n)$ vs $O(g(n))$'

# Graficar funciones
ax.plot(n, f_values, label='$C(n) = n^3 + 2n^2 + n + 5$', color='#1565C0')
ax.plot(n, g_values, label=label_text, color='#B71C1C')

# Línea vertical en n₀
if n0:
    ax.axvline(
        x=n0,
        color='#2E7D32',
        linestyle='--',
        label=rf'$n_0 = {format_n0(n0)}$'
    )

    ax.fill_between(n, f_values, g_values, where=(f_values <= g_values), interpolate=True, color='#B71C1C', alpha=0.3)

# Escala log-log
ax.set_xscale('log')
ax.set_yscale('log')

# Etiquetas y título
ax.set_xlabel('Tamaño de la entrada ($n$)')
ax.set_ylabel('$O(g(n))$')
ax.set_title(title_text)
ax.legend(loc='upper left', bbox_to_anchor=(0.07, 1))
ax.grid(True, which="both", ls="--", linewidth=0.5)

# Mostrar la gráfica
plt.show()

# Guardar la gráfica como archivo PNG
fig.savefig(GRAPH_DIR / 'comparacion_big_O_2.png', bbox_inches='tight')
```

## Ejemplo del libro


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import brentq

plt.style.use('default')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['savefig.edgecolor'] = 'white'

plt.rcParams['figure.dpi'] = 500
plt.rcParams['savefig.dpi'] = 500

plt.rcParams['savefig.pad_inches'] = 0.05

plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
```

```python
# Definir las funciones
def C(n):
    return n**3 + 2*n**2 + n + 5

def g(n):
    return n**3

# Constante c para Omega
c = 0.5

# Función diferencia
def diff(n):
    return C(n) - c * g(n)

# Evaluar extremos
f1 = diff(1)
f2 = diff(1e6)

# Determinar si hay cruce de signo
try:
    if f1 * f2 < 0:
        n0 = brentq(diff, 1, 1e6)
    else:
        n0 = 0
except ValueError:
    n0 = 0


# Crear valores de n
n1 = np.linspace(1, 1e4, 5000)
n2 = np.linspace(1e4, 1e20, 5000)
n = np.concatenate((n1, n2))

# Evaluar funciones
f_values = C(n)
g_values = c * g(n)

# Crear la gráfica
fig, ax = plt.subplots(figsize=(8, 3))

# Etiquetas
label_text = f'$c \\cdot g(n) = {c}n^3$'
title_text = '$C(n)$ vs $\\Omega(g(n))$'

# Graficar
ax.plot(n, f_values, label='$C(n) = n^3 + 2n^2 + n + 5$', color='#1565C0')
ax.plot(n, g_values, label=label_text, color='#EF6C00')
ax.axvline(x=n0, color='#2E7D32', linestyle='--', label=f'$n_0 = {format_n0(n0)}$')
ax.fill_between(n, g_values, f_values, where=(f_values >= g_values), color='#1565C0', alpha=0.25)

# Etiquetas
ax.set_xlabel('Tamaño de la entrada ($n$)')
ax.set_ylabel('$\\Omega(g(n))$')
ax.set_title(title_text)
ax.legend(loc='upper left', bbox_to_anchor=(0.06, 1))
ax.grid(True)

# Mostrar y guardar
plt.show()
fig.savefig(GRAPH_DIR / 'comparacion_big_omega.png', bbox_inches='tight')
```

## 5 notacion theta


## Ejemplo del libro


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import brentq

plt.style.use('default')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['savefig.edgecolor'] = 'white'

plt.rcParams['figure.dpi'] = 500
plt.rcParams['savefig.dpi'] = 500

plt.rcParams['savefig.pad_inches'] = 0.05

plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
```

```python
# Definir funciones
def C(n):
    return n**3 + 2*n**2 + n + 5

def g(n):
    return n**3

# Constantes
c1 = 0.5
c2 = 2

# Funciones diferencia para encontrar n0
def diff_lower(n):
    return C(n) - c1 * g(n)

def diff_upper(n):
    return c2 * g(n) - C(n)

# Encontrar los n0 correspondientes
try:
    n0_lower = brentq(diff_lower, 1, 1e6)
except ValueError:
    n0_lower = 0
try:
    n0_upper = brentq(diff_upper, 1, 1e6)
except ValueError:
    n0_upper = 0

# Tomamos el mayor de ambos como verdadero n0 para Big-Theta
n0_theta = max(n0_lower, n0_upper)

# Rango de n combinado
n1 = np.linspace(1, 1e4, 5000)
n2 = np.linspace(1e4, 1e20, 5000)
n = np.concatenate((n1, n2))

# Evaluar funciones
f_values = C(n)
g_lower = c1 * g(n)
g_upper = c2 * g(n)

# Crear gráfica
fig, ax = plt.subplots(figsize=(8, 3))

# Etiquetas
title_text = '$C(n)$ vs $\\Theta(g(n))$'
label_c1 = f'$c_1 \\cdot g(n) = {c1}n^3$'
label_c2 = f'$c_2 \\cdot g(n) = {c2}n^3$'

# Graficar funciones
ax.plot(n, f_values, label='$C(n) = n^3 + 2n^2 + n + 5$', color='#1565C0')
ax.plot(n, g_lower, label=label_c1, color='#EF6C00')
ax.plot(n, g_upper, label=label_c2, color='#B71C1C')

# Línea vertical en n0
ax.axvline(x=n0_theta, color='#2E7D32', linestyle='--', label=rf'$n_0 = {format_n0(n0_theta)}$')

# Rellenar región válida para Theta
ax.fill_between(n, g_lower, g_upper, where=(f_values >= g_lower) & (f_values <= g_upper), color='#1565C0', alpha=0.25)

# Etiquetas
ax.set_xlabel('Tamaño de la entrada ($n$)')
ax.set_ylabel('$\\Theta(g(n))$')
ax.set_title(title_text)
ax.legend(loc='upper left', bbox_to_anchor=(0.06, 1))
ax.grid(True)

# Mostrar y guardar
plt.show()
fig.savefig(GRAPH_DIR / 'comparacion_big_theta.png', bbox_inches='tight')
```
