```python
import time
import tracemalloc
from scipy.optimize import curve_fit
import numpy as np
from util import graficar_complejidad, modelo_cuadratico
import matplotlib.pyplot as plt

plt.style.use('default')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['savefig.edgecolor'] = 'white'

plt.rcParams['figure.dpi'] = 500
plt.rcParams['savefig.dpi'] = 500
plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
```

```python
def medir_tiempo_matriz_saltos(func, sizes, n_iter):
    tiempos = np.zeros(len(sizes))
    for i, n in enumerate(sizes):
        acumulado = 0.0
        for _ in range(n_iter):
            inicio = time.perf_counter()
            func(n, n)
            acumulado += time.perf_counter() - inicio
        tiempos[i] = acumulado / n_iter
    return tiempos

def medir_memoria_matriz_saltos(func, sizes, n_iter):
    espacios = np.zeros(len(sizes), dtype=int)
    for i, n in enumerate(sizes):
        acumulado = 0
        for _ in range(n_iter):
            tracemalloc.start()
            func(n, n)
            _, peak = tracemalloc.get_traced_memory()
            tracemalloc.stop()
            acumulado += peak
        espacios[i] = acumulado / n_iter
    return espacios

def recorrer_matriz_vacia(m, n):
    matriz = [[0 for _ in range(n)] for _ in range(m)]
    for i in range(m):
        for j in range(0, n, 2):
            pass
```

```python
n_ejecuciones = 200
sizes = np.arange(1, 200, 10)
x = sizes
```

```python
tiempos = medir_tiempo_matriz_saltos(recorrer_matriz_vacia, sizes, n_ejecuciones)
params_tiempo = curve_fit(modelo_cuadratico, sizes**2, tiempos)[0]
tiempos_ajustados = modelo_cuadratico(sizes**2, *params_tiempo)

graficar_complejidad(
    x=sizes,
    y_experimental=tiempos,
    y_teorico=tiempos_ajustados,
    nombre_archivo="recorrer_matriz_vacia_tiempo.png",
    ylabel="Tiempo de ejecución [s]",
    funcion="T(n)",
    xlabel="Cantidad de datos de entrada ($m \\cdot n$)"
)
```

```python

recursos = medir_memoria_matriz_saltos(recorrer_matriz_vacia, sizes, n_ejecuciones)
params_memoria = curve_fit(modelo_cuadratico, sizes**2, recursos)[0]
recursos_ajustados = modelo_cuadratico(sizes**2, *params_memoria)

graficar_complejidad(
    x=sizes,
    y_experimental=recursos,
    y_teorico=recursos_ajustados,
    nombre_archivo="recorrer_matriz_vacia_espacio.png",
    ylabel="Consumo de memoria [bytes]",
    funcion="S(n)",
    xlabel="Cantidad de datos de entrada ($m \\cdot n$)"
)
```
