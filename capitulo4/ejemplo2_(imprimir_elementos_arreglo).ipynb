```python
import time
import tracemalloc

import numpy as np
from scipy.optimize import curve_fit

from util import graficar_complejidad, modelo_constante, modelo_lineal
import matplotlib.pyplot as plt

plt.style.use('default')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['savefig.edgecolor'] = 'white'

plt.rcParams['figure.dpi'] = 500
plt.rcParams['savefig.dpi'] = 500
plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
```

```python
def medir_tiempo_arreglos(func, sizes, n_iter):
    tiempos = np.zeros(len(sizes))

    for i, n in enumerate(sizes):
        acumulado = 0.0
        for _ in range(n_iter):
            arr = np.random.randint(0, 1000, size=n)
            inicio = time.perf_counter()
            func(arr)
            acumulado += time.perf_counter() - inicio
        tiempos[i] = acumulado / n_iter

    return tiempos


def medir_memoria_arreglos(func, sizes, n_iter):
    espacios = np.zeros(len(sizes), dtype=int)

    for i, n in enumerate(sizes):
        acumulado = 0
        for _ in range(n_iter):
            arr = np.random.randint(0, 1000, size=n)
            tracemalloc.start()
            func(arr)
            _, peak = tracemalloc.get_traced_memory()
            tracemalloc.stop()
            acumulado += peak
        espacios[i] = acumulado / n_iter

    return espacios

def imprimir_elementos(arr):
    for x in arr:
        _ = x
```

```python
n_ejecuciones = 200
sizes = np.arange(1, 10000, 100)
x = sizes
```

```python
tiempos = medir_tiempo_arreglos(imprimir_elementos, sizes, n_ejecuciones)
params_tiempo = curve_fit(modelo_lineal, sizes, tiempos)[0]
tiempos_ajustados = modelo_lineal(sizes, *params_tiempo)

graficar_complejidad(
    x=sizes,
    y_experimental=tiempos,
    y_teorico=tiempos_ajustados,
    nombre_archivo='ejemplo_imprimir_elementos_tiempo.png',
    ylabel='Tiempo de ejecución [s]',
    funcion='T(n)'
)
```

```python
recursos = medir_memoria_arreglos(imprimir_elementos, sizes, n_ejecuciones)
params_memoria = curve_fit(modelo_constante, sizes, recursos)[0]
recursos_ajustados = modelo_constante(sizes, *params_memoria)

graficar_complejidad(
    x=sizes,
    y_experimental=recursos,
    y_teorico=recursos_ajustados,
    nombre_archivo='ejemplo_imprimir_elementos_espacio.png',
    ylabel='Consumo de memoria [bytes]',
    funcion='S(n)'
)
```
