```python
import time
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import sys
from pathlib import Path
plt.style.use('default')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['savefig.edgecolor'] = 'white'

plt.rcParams['figure.dpi'] = 500
plt.rcParams['savefig.dpi'] = 500
```

Medición de tiempo de ejecución para algoritmo de búsqueda binaria dado un arreglo ordenado

```python
def busqueda_binaria_recursiva(lista, objetivo, bajo, alto):
    if bajo > alto:
        return -1

    medio = (bajo + alto) // 2

    if lista[medio] == objetivo:
        return medio
    elif lista[medio] < objetivo:
        return busqueda_binaria_recursiva(lista, objetivo, medio + 1, alto)
    else:
        return busqueda_binaria_recursiva(lista, objetivo, bajo, medio - 1)

def medir_tiempo_busqueda_binaria_recursiva(lista, repeticiones=50):
    tiempos = np.zeros(repeticiones)
    objetivo = lista[-1]
    for i in range(repeticiones):
        inicio = time.perf_counter()
        busqueda_binaria_recursiva(lista, objetivo, 0, len(lista) - 1)
        fin = time.perf_counter()
        tiempos[i] = fin - inicio
    return np.average(tiempos)
```

Gráfica de n vs t[segundos]

```python
listas = np.arange(1, 10000000, 50000)
tiempos_promedio_busqueda = np.zeros(len(listas))

for i, n in enumerate(listas):
    lista = list(range(n))
    tiempos_promedio_busqueda[i] = medir_tiempo_busqueda_binaria_recursiva(lista)

log_listas = np.log2(listas)
slope_log, intercept_log, r_value_log, p_value_log, std_err_log = stats.linregress(log_listas, tiempos_promedio_busqueda)
logaritmica_ajustada = slope_log * log_listas + intercept_log

fig, ax1 = plt.subplots(1, 1, figsize=(8, 4))
ax1.plot(listas, tiempos_promedio_busqueda, label='Función de complejidad experimental')
ax1.plot(listas, logaritmica_ajustada, label=f'Función de complejidad teórica $T(n) = \log_2(n)$', linestyle='dotted', color='red')

ax1.set_xlabel('Tamaño de la entrada ($n$)')
ax1.set_ylabel('Tiempo de ejecución promedio [s]')
ax1.set_xlim(left=0)
ax1.set_ylim(bottom=0)
ax1.set_title(f'T(n) teórico vs T(n) calculado')
ax1.legend(loc='lower right')
ax1.grid(True)

ax1.xaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
ax1.yaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
ax1.ticklabel_format(axis='x', style='sci', scilimits=(0, 0))
ax1.ticklabel_format(axis='y', style='sci', scilimits=(0, 0))

plt.tight_layout()
plt.show()
Path("graficas").mkdir(parents=True, exist_ok=True)
fig.savefig('graficas/complejidad_logaritmica.png', bbox_inches="tight", pad_inches=.05)
```
