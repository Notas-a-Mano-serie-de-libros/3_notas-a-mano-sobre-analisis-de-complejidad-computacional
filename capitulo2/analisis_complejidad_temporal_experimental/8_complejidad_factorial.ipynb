```python
import time
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import math
from pathlib import Path
plt.style.use('default')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['savefig.edgecolor'] = 'white'

plt.rcParams['figure.dpi'] = 500
plt.rcParams['savefig.dpi'] = 500
```

```python

def permutar_lista(lista):
    if len(lista) == 0:
        return []
    if len(lista) == 1:
        return [lista]

    permutaciones = []

    for i in range(len(lista)):

        actual = lista[i]

        restante = lista[:i] + lista[i+1:]

        for p in permutar_lista(restante):
            permutaciones.append([actual] + p)

    return permutaciones

def medir_tiempo_promedio_permutaciones(n, repeticiones=5):
    tiempos = np.zeros(repeticiones)
    for i in range(repeticiones):
        lista = list(range(1, n + 1))
        inicio = time.perf_counter()
        permutar_lista(lista)
        fin = time.perf_counter()
        tiempos[i] = fin - inicio
    return np.average(tiempos)
```

```python

listas = np.arange(1, 11, 1)

tiempos_promedio_permutaciones = np.zeros(len(listas))

for i, n in enumerate(listas):
    tiempos_promedio_permutaciones[i] = medir_tiempo_promedio_permutaciones(n)

factorial_listas = [math.factorial(n) for n in listas]
slope_factorial, intercept_factorial, r_value_factorial, p_value_factorial, std_err_factorial = stats.linregress(factorial_listas, tiempos_promedio_permutaciones)

factorial_ajustada_tiempos = slope_factorial * np.array(factorial_listas) + intercept_factorial

fig, ax1 = plt.subplots(1, 1, figsize=(8, 4))

ax1.plot(listas, tiempos_promedio_permutaciones, label='Función de complejidad experimental')

ax1.plot(listas, factorial_ajustada_tiempos, label=f'Función de complejidad teórica $T(n) = n!$', linestyle='dotted', color='red')

ax1.set_xlabel('Tamaño de la entrada ($n$)')
ax1.set_ylabel('Tiempo de ejecución promedio [s]')
ax1.set_xlim(left=1)
ax1.set_ylim(bottom=-.5)
ax1.set_title(f'T(n) teórico vs T(n) calculado')
ax1.legend(loc='upper left')
ax1.grid(True)

ax1.xaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
ax1.yaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))

plt.tight_layout()
plt.show()

Path("graficas").mkdir(parents=True, exist_ok=True)
fig.savefig('graficas/complejidad_factorial.png', bbox_inches="tight", pad_inches=.05)
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(8, 4))

listas_extendidas = np.arange(1, 71, 1, dtype=np.int64)
funcion_teorica_factorial = np.array([math.factorial(int(n)) for n in listas_extendidas], dtype=np.float64)

ax1.plot(listas_extendidas, funcion_teorica_factorial, label=f'Función de complejidad teórica $T(n) = n!$', linestyle='dotted', color='red')

ax1.set_xlabel('Tamaño de la entrada ($n$)')
ax1.set_ylabel('Valor teórico de $T(n)$')

ax1.set_xlim(left=1, right=70)
ax1.set_yscale('log')

ax1.xaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
ax1.yaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))

ax1.set_title(f'T(n) teórico (Rango extendido hasta $70! \\approx 10^{{100}}$)')
ax1.legend(loc='lower right')
ax1.grid(True)

plt.tight_layout()
plt.show()
Path("graficas").mkdir(parents=True, exist_ok=True)
fig.savefig('graficas/complejidad_factorial_2.png', bbox_inches="tight", pad_inches=.05)
```
