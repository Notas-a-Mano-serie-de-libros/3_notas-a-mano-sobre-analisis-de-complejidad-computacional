```python
import time
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from pathlib import Path
plt.style.use('default')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['savefig.edgecolor'] = 'white'

plt.rcParams['figure.dpi'] = 500
plt.rcParams['savefig.dpi'] = 500
```

```python

def recorrer_matriz(n):
    matriz = np.random.randint(0, 100, size=(n, n))
    suma = 0
    for i in range(n):
        for j in range(n):
            suma += matriz[i][j]
    return suma

def medir_tiempo_promedio_cuadratico(n, repeticiones=10):
    tiempos = np.zeros(repeticiones)
    for i in range(repeticiones):
        inicio = time.perf_counter()
        recorrer_matriz(n)
        fin = time.perf_counter()
        tiempos[i] = fin - inicio
    return np.average(tiempos)
```

```python

listas = np.arange(1, 1000, 10)

tiempos_promedio_cuadratico = np.zeros(len(listas))

for i, n in enumerate(listas):
    tiempos_promedio_cuadratico[i] = medir_tiempo_promedio_cuadratico(n)

cuadratica_listas = listas ** 2
slope_cuadratica, intercept_cuadratica, r_value_cuadratica, p_value_cuadratica, std_err_cuadratica = stats.linregress(cuadratica_listas, tiempos_promedio_cuadratico)

cuadratica_ajustada_tiempos = slope_cuadratica * cuadratica_listas + intercept_cuadratica

fig, ax1 = plt.subplots(1, 1, figsize=(8, 4))

ax1.plot(listas, tiempos_promedio_cuadratico, label='Función de complejidad experimental')

ax1.plot(listas, cuadratica_ajustada_tiempos, label=f'Función de complejidad teórica $T(n) = n^2$', linestyle='dotted', color='red')

ax1.set_xlabel('Tamaño de la entrada ($n$)')
ax1.set_ylabel('Tiempo de ejecución promedio [s]')
ax1.set_xlim(left=0)
ax1.set_ylim(bottom=0)
ax1.set_title(f'T(n) teórico vs T(n) calculado')
ax1.legend(loc='upper left')
ax1.grid(True)

ax1.xaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
ax1.yaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
ax1.ticklabel_format(axis='x', style='sci', scilimits=(0, 0))
ax1.ticklabel_format(axis='y', style='sci', scilimits=(0, 0))

plt.tight_layout()
plt.show()

Path("graficas").mkdir(parents=True, exist_ok=True)
fig.savefig('graficas/complejidad_cuadratica.png', bbox_inches="tight", pad_inches=.05)
```
