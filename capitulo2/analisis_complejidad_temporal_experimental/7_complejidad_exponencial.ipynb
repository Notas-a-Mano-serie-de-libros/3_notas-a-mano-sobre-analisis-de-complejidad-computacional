```python
import time
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from pathlib import Path
plt.style.use('default')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['savefig.edgecolor'] = 'white'

plt.rcParams['figure.dpi'] = 500
plt.rcParams['savefig.dpi'] = 500
```

```python

def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

def medir_tiempo_promedio_fibonacci(n, repeticiones=5):
    tiempos = np.zeros(repeticiones)
    for i in range(repeticiones):
        inicio = time.perf_counter()
        fibonacci(n)
        fin = time.perf_counter()
        tiempos[i] = fin - inicio
    return np.average(tiempos)
```

```python

listas = np.arange(1, 40, 1)

tiempos_promedio_fibonacci = np.zeros(len(listas))

for i, n in enumerate(listas):
    tiempos_promedio_fibonacci[i] = medir_tiempo_promedio_fibonacci(n)

exponencial_listas = 2 ** listas
slope_exponencial, intercept_exponencial, r_value_exponencial, p_value_exponencial, std_err_exponencial = stats.linregress(exponencial_listas, tiempos_promedio_fibonacci)

exponencial_ajustada_tiempos = slope_exponencial * exponencial_listas + intercept_exponencial

fig, ax1 = plt.subplots(1, 1, figsize=(8, 4))

ax1.plot(listas, tiempos_promedio_fibonacci, label='Función de complejidad experimental')

ax1.plot(listas, exponencial_ajustada_tiempos, label=f'Función de complejidad teórica $T(n) = 2^n$', linestyle='dotted', color='red')

ax1.set_xlabel('Tamaño de la entrada ($n$)')
ax1.set_ylabel('Tiempo de ejecución promedio [s]')
ax1.set_xlim(left=0)
ax1.set_ylim(bottom=0)
ax1.set_title(f'T(n) teórico vs T(n) calculado')
ax1.legend(loc='upper left')
ax1.grid(True)

ax1.xaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
ax1.yaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))

plt.tight_layout()
plt.show()

Path("graficas").mkdir(parents=True, exist_ok=True)
fig.savefig('graficas/complejidad_exponencial.png', bbox_inches="tight", pad_inches=.05)
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(8, 4))

listas_extendidas = np.arange(1, 334, 1, dtype=np.float64)
funcion_teorica_exponencial = np.power(2.0, listas_extendidas)

ax1.plot(listas_extendidas, funcion_teorica_exponencial, label=f'Función de complejidad teórica $T(n) = 2^n$', linestyle='dotted', color='red')

ax1.set_xlabel('Tamaño de la entrada ($n$)')
ax1.set_ylabel('Valor teórico de $T(n)$')

ax1.set_xlim(left=0, right=333)
ax1.set_yscale('log')

ax1.xaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
ax1.yaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
ax1.ticklabel_format(axis='x', style='sci', scilimits=(0, 0))

ax1.set_title(f'T(n) teórico (Rango extendido hasta $2^{{333}} \\approx 10^{{100}}$)')
ax1.legend(loc='lower right')
ax1.grid(True)

plt.tight_layout()
plt.show()
Path("graficas").mkdir(parents=True, exist_ok=True)
fig.savefig('graficas/complejidad_exponencial_2.png', bbox_inches="tight", pad_inches=.05)
```
