```python
import time
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from pathlib import Path

plt.style.use('default')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['savefig.edgecolor'] = 'white'

plt.rcParams['figure.dpi'] = 500
plt.rcParams['savefig.dpi'] = 500
```

```python

def heapsort(arr):
    n = len(arr)

    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)

def heapify(arr, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n and arr[left] > arr[largest]:
        largest = left

    if right < n and arr[right] > arr[largest]:
        largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def medir_tiempo_promedio_heapsort(lista, repeticiones=10):
    tiempos = np.zeros(repeticiones)
    for i in range(repeticiones):
        inicio = time.perf_counter()
        heapsort(lista)
        fin = time.perf_counter()
        tiempos[i] = fin - inicio
    return np.average(tiempos)
```

```python

listas = np.arange(1, 10000, 50)

tiempos_promedio_heapsort = np.zeros(len(listas))

for i, n in enumerate(listas):
    lista = np.random.randint(0, n, n)
    tiempos_promedio_heapsort[i] = medir_tiempo_promedio_heapsort(lista)

log_lineal_listas = listas * np.log2(listas)
slope_loglineal, intercept_loglineal, r_value_loglineal, p_value_loglineal, std_err_loglineal = stats.linregress(log_lineal_listas, tiempos_promedio_heapsort)

loglineal_ajustada_tiempos = slope_loglineal * log_lineal_listas + intercept_loglineal

fig, ax1 = plt.subplots(1, 1, figsize=(8, 4))

ax1.plot(listas, tiempos_promedio_heapsort, label='Función de complejidad experimental')

ax1.plot(listas, loglineal_ajustada_tiempos, label=f'Función de complejidad teórica $T(n) = n \log_2(n)$', linestyle='dotted', color='red')

ax1.set_xlabel('Tamaño de la entrada ($n$)')
ax1.set_ylabel('Tiempo de ejecución promedio [s]')
ax1.set_xlim(left=0)
ax1.set_ylim(bottom=0)
ax1.set_title(f'T(n) teórico vs T(n) calculado')
ax1.legend(loc='upper left')
ax1.grid(True)

ax1.xaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
ax1.yaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
ax1.ticklabel_format(axis='x', style='sci', scilimits=(0, 0))
ax1.ticklabel_format(axis='y', style='sci', scilimits=(0, 0))

plt.tight_layout()
plt.show()

Path("graficas").mkdir(parents=True, exist_ok=True)
fig.savefig('graficas/complejidad_log_lineal.png', bbox_inches="tight", pad_inches=.05)
```
