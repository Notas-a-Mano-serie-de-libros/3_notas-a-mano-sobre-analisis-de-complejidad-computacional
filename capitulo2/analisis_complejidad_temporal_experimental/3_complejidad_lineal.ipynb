```python
import time
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from pathlib import Path

plt.style.use('default')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['savefig.edgecolor'] = 'white'

plt.rcParams['figure.dpi'] = 500
plt.rcParams['savefig.dpi'] = 500
```

```python
def medir_tiempo_buscar_elemento_arreglo(lista, elemento, repeticiones=10):
    tiempos = np.zeros(repeticiones)
    for i in range(repeticiones):
        inicio = time.perf_counter()
        for current in lista :
            if elemento == current: 
                break
        fin = time.perf_counter()
        tiempos[i] = fin - inicio
    return np.average(tiempos)
```

```python
listas = np.arange(1, 1000000, 2500)
tiempos_promedio = np.zeros(len(listas))

for i, n in enumerate(listas):
    lista = list(range(n))
    tiempos_promedio[i] = medir_tiempo_buscar_elemento_arreglo(lista, lista[-1])

slope, intercept, r_value, p_value, std_err = stats.linregress(listas, tiempos_promedio)
lineal_ajustada = np.array(listas) * slope + intercept

fig, ax1 = plt.subplots(1, 1, figsize=(8, 4))
ax1.plot(listas, tiempos_promedio, label='Función de complejidad experimental')
ax1.plot(listas, lineal_ajustada, label=f'Función de complejidad teórica $T(n) = n$', linestyle='dotted', color='red')

ax1.set_xlabel('Tamaño de la entrada ($n$)')
ax1.set_ylabel('Tiempo de ejecución promedio [s]')
ax1.set_xlim(left=0)
ax1.set_ylim(bottom=0)
ax1.set_title(f'T(n) teórico vs T(n) calculado')
ax1.legend(loc='lower right')
ax1.grid(True)

ax1.xaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
ax1.yaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
ax1.ticklabel_format(axis='x', style='sci', scilimits=(0, 0))
ax1.ticklabel_format(axis='y', style='sci', scilimits=(0, 0))

plt.tight_layout()
plt.show()
Path("graficas").mkdir(parents=True, exist_ok=True)
fig.savefig('graficas/complejidad_lineal.png', bbox_inches="tight", pad_inches=.05)
```
