# Complejidad constante

La complejidad constante describe operaciones cuyo costo permanece estable frente a cambios en el tamaño de la entrada. El tamaño de la entrada se representa con $n$ y suele indicar la cantidad de elementos disponibles para procesar: por ejemplo, la longitud de una lista, la cantidad de registros en una tabla o el número de posiciones de un arreglo.

Una operación de costo constante realiza una cantidad fija de acciones elementales. Esa cantidad puede ser pequeña o grande, dependiendo de la operación concreta y de la máquina que la ejecuta, pero conserva una característica esencial: al aumentar $n$, el procedimiento ejecutado mantiene la misma estructura y la misma cantidad básica de trabajo.

Este tipo de comportamiento aparece en instrucciones puntuales como leer una variable, asignar un valor, comparar dos datos simples o acceder a una posición conocida dentro de una estructura con acceso directo.


## Forma teórica

Para una entrada de tamaño $n$, una función de costo constante puede expresarse como:

$$T(n) = c$$

donde $T(n)$ representa el costo de ejecutar la operación sobre una entrada de tamaño $n$, y $c$ representa una cantidad fija de trabajo. Ese valor puede interpretarse como tiempo, número de instrucciones, número de accesos a memoria o cualquier unidad de medición definida para el análisis.

La expresión $T(n)=c$ indica que el costo se mantiene en el mismo nivel para diferentes valores de $n$. Si $n=10$, $n=10.000$ o $n=10.000.000$, el modelo teórico asigna el mismo costo a la operación estudiada.

Esto ocurre cuando el algoritmo llega directamente a la información requerida. La operación evaluada tiene una ruta de ejecución fija: recibe la entrada, identifica la posición o dato requerido, realiza la acción y termina. El número total de elementos disponibles modifica el tamaño de la estructura, pero la operación puntual sigue ocupando el mismo lugar dentro del proceso.

En una gráfica, una función constante se representa mediante una línea horizontal. La altura de esa línea corresponde al valor de $c$. Si la línea está más arriba, la operación tarda más en términos absolutos; si está más abajo, tarda menos. En ambos casos, el rasgo importante es la ausencia de crecimiento sostenido al aumentar $n$.

## Ejemplo: acceder a una posición de un arreglo

El ejemplo usado es el acceso a una posición específica de una lista o arreglo:

```python
_ = lista[indice]
```

En esta instrucción, `lista` representa la estructura de datos, `indice` representa la posición que se desea consultar y `_` recibe el valor leído. La operación consiste en obtener el elemento almacenado en una ubicación conocida.

El acceso por índice tiene comportamiento constante porque la posición del elemento se calcula directamente. La ejecución requiere conocer la referencia inicial de la estructura y el desplazamiento asociado al índice. Con esa información, el entorno de ejecución puede ubicar el elemento solicitado mediante una operación directa de acceso.

La lista puede contener muchos elementos, pero el acceso a `lista[indice]` consulta una sola posición. La operación realizada para leer el elemento central de una lista de 100 posiciones tiene la misma forma que la operación realizada para leer el elemento central de una lista de 1.000.000 de posiciones: se identifica un índice válido y se recupera el dato ubicado en esa posición.

La diferencia entre listas pequeñas y grandes aparece en otras operaciones, como construir la lista, recorrerla completa, copiarla o buscar un valor desconocido. En cambio, cuando el índice ya está determinado, el acceso se concentra en una única ubicación.


## Función de medición

La siguiente celda define una función para medir el tiempo de acceso a una posición del arreglo. La función recibe una lista y una cantidad de repeticiones. Para cada repetición, calcula el instante inicial, accede al elemento ubicado en la mitad de la lista, calcula el instante final y almacena la diferencia.

El índice seleccionado es:

$$indice = rac{len(lista)}{2}$$

En el código se usa división entera mediante `//`, de modo que el índice siempre corresponde a una posición válida. Elegir la mitad de la lista ayuda a evitar casos extremos como la primera o la última posición, aunque para el acceso por índice esas posiciones conservan el mismo tipo de operación.

La medición se repite varias veces porque los tiempos individuales pueden variar por factores externos: carga del sistema, administración de memoria, planificación del procesador, estado de la caché, interrupciones del sistema operativo o pequeñas fluctuaciones del entorno de ejecución. El promedio reduce el efecto de mediciones aisladas y ofrece una estimación más estable del costo observado.


````python
#@title Medición de tiempo { display-mode: "form" }

def medir_tiempo_acceso_arreglo(lista, repeticiones=10):
    tiempos = np.zeros(repeticiones)
    indice = len(lista) // 2
    for i in range(repeticiones):
        inicio = time.perf_counter()
        _ = lista[indice]
        fin = time.perf_counter()
        tiempos[i] = fin - inicio
    return np.average(tiempos)
````

## Simulación experimental

La simulación crea listas de distintos tamaños usando una secuencia de valores de $n$. Para cada tamaño se construye una lista con `range(n)` y se mide el tiempo promedio de acceder al elemento ubicado en la mitad.

El procedimiento experimental sigue estos pasos:

1. Se define un conjunto de tamaños de entrada.
2. Para cada tamaño $n$, se construye una lista con $n$ elementos.
3. Se calcula el índice central de esa lista.
4. Se mide varias veces el acceso a `lista[indice]`.
5. Se almacena el tiempo promedio obtenido para ese tamaño.
6. Se calcula un promedio general de todas las mediciones.
7. Se dibuja una línea horizontal que sirve como referencia del modelo constante.

En la gráfica se muestran dos elementos. La curva experimental representa los tiempos medidos para cada tamaño de lista. La línea punteada roja representa una referencia constante, construida a partir del promedio general de las mediciones.

La línea experimental puede presentar oscilaciones porque el tiempo medido es extremadamente pequeño. Cuando una operación tarda fracciones muy reducidas de segundo, cualquier actividad del sistema puede afectar la medición. Por esa razón, la interpretación debe concentrarse en la tendencia general: el tiempo permanece alrededor de un nivel estable mientras $n$ aumenta.


````python
#@title Cálculo de complejidad temporal teórica { display-mode: "form" }

listas = np.arange(1, 10000000, 50000)
tiempos_promedio_acceso = np.zeros(len(listas))

for i, n in enumerate(listas):
    lista = list(range(n))
    tiempos_promedio_acceso[i] = medir_tiempo_acceso_arreglo(lista)

promedio_acceso = np.mean(tiempos_promedio_acceso)
constante_ajustada = np.full(len(listas), promedio_acceso)

fig, ax1 = plt.subplots(1, 1, figsize=(8, 4))

ax1.plot(listas, tiempos_promedio_acceso, label='Función de complejidad experimental')
ax1.plot(listas, constante_ajustada, label=f'Función de complejidad teórica T(n) = 1', linestyle='dotted', color='red')

ax1.set_xlabel('Tamaño de la entrada ($n$)')
ax1.set_ylabel('Tiempo de ejecución promedio [s]')
ax1.set_xlim(left=0)
ax1.set_ylim(bottom=0)
ax1.set_title(f'T(n) teórico vs T(n) calculado')
ax1.legend(loc='upper right')
ax1.grid(True)

ax1.xaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
ax1.yaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
ax1.ticklabel_format(axis='x', style='sci', scilimits=(0, 0))
ax1.ticklabel_format(axis='y', style='sci', scilimits=(0, 0))

plt.tight_layout()
plt.show()
fig.savefig('graficas/complejidad_constante.png', bbox_inches="tight", pad_inches=.05)
````

## Interpretación de la primera gráfica

La primera gráfica compara la medición experimental con la referencia teórica constante. El eje horizontal representa el tamaño de la lista, es decir, la cantidad de elementos disponibles. El eje vertical representa el tiempo promedio medido para acceder a una posición.

La curva experimental muestra pequeñas variaciones entre una medición y otra. Estas variaciones son normales en mediciones de tiempo real, especialmente cuando la operación observada es tan breve. Un punto puede quedar un poco más alto por una interrupción del sistema, por cambios en la memoria caché o por el costo de crear y administrar objetos durante la ejecución.

La referencia constante aparece como una línea horizontal. Su función es mostrar el comportamiento esperado cuando el costo permanece estable. La comparación relevante consiste en observar si la curva experimental forma una pendiente creciente a medida que $n$ aumenta.

Los resultados muestran que el tiempo de acceso se mantiene alrededor de un rango pequeño. La lista crece de manera significativa, pero el acceso por índice conserva una duración parecida. Esa estabilidad respalda la idea de que la operación consultada tiene costo constante.

El experimento también permite distinguir entre el costo de construir la lista y el costo de acceder a un elemento. La lista se crea dentro del ciclo experimental, pero la medición temporal se toma únicamente alrededor de la instrucción de acceso. Por eso, la gráfica describe el costo de leer una posición conocida, no el costo total de crear o llenar la estructura.


## Variación alrededor del promedio

La segunda visualización transforma los datos para observar mejor las fluctuaciones. A cada medición se le resta el promedio general:

$$variación(n) = tiempo\_medido(n) - promedio\_general$$

Después de esta transformación, el valor cero representa el promedio. Los puntos ubicados por encima de cero indican mediciones ligeramente más lentas que el promedio. Los puntos ubicados por debajo de cero indican mediciones ligeramente más rápidas.

Esta gráfica es útil porque separa la forma general del comportamiento respecto a pequeñas desviaciones. Si los tiempos crecieran de forma sostenida con el tamaño de la lista, las diferencias respecto al promedio tenderían a organizarse en una dirección clara. En el caso de una operación constante, las diferencias suelen distribuirse alrededor de cero, con picos aislados y variaciones pequeñas.

La presencia de picos individuales debe interpretarse como ruido experimental. El acceso por índice es tan rápido que la medición queda expuesta a detalles del entorno: actividad de otros procesos, calentamiento de caché, recolección de memoria, asignaciones previas y precisión del temporizador. Lo importante es que las desviaciones no forman una tendencia creciente asociada al tamaño de la entrada.


````python
#@title Variación respecto al promedio { display-mode: "form" }

fig, ax1 = plt.subplots(1, 1, figsize=(8, 4))

ax1.plot(listas, tiempos_promedio_acceso - promedio_acceso, label='Función de complejidad experimental')
ax1.plot(listas, constante_ajustada - promedio_acceso, label=f'Función de complejidad teórica T(n) = 1', linestyle='dotted', color='red')

ax1.set_xlabel('Tamaño de la entrada ($n$)')
ax1.set_ylabel('Tiempo de ejecución promedio [s]')
ax1.set_xlim(left=0)
ax1.set_ylim(-1, 1)
ax1.set_title(f'T(n) teórico vs T(n) calculado')
ax1.legend(loc='upper right')
ax1.grid(True)

ax1.xaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
ax1.ticklabel_format(axis='x', style='sci', scilimits=(0, 0))

plt.tight_layout()
plt.show()
Path("graficas").mkdir(parents=True, exist_ok=True)
fig.savefig('graficas/complejidad_constante_3.png', bbox_inches="tight", pad_inches=.05)
````

## Conclusión

El experimento confirma que acceder a una posición conocida de un arreglo tiene comportamiento constante. La razón principal es que la operación consulta una ubicación determinada mediante un índice; el procedimiento ejecutado no requiere inspeccionar cada elemento de la estructura.

El tamaño de la lista cambia durante la simulación, pero la acción medida conserva la misma forma: seleccionar una posición y leer el valor almacenado allí. Por eso, los tiempos observados se mantienen alrededor de un mismo nivel y las variaciones se explican mejor como ruido de medición que como crecimiento del costo.

Este resultado es importante porque muchas estructuras y algoritmos dependen de accesos puntuales. Cuando una operación básica tiene costo estable, puede usarse como componente confiable dentro de procesos más grandes. Aun así, conviene diferenciar el acceso directo de otras tareas sobre listas: recorrer todos los elementos, buscar un valor desconocido, insertar en ciertas posiciones o construir una lista completa pueden tener comportamientos diferentes.

La idea práctica queda resumida así: si el índice ya se conoce y la estructura permite acceso directo, leer el elemento correspondiente mantiene un costo estable frente al crecimiento de la entrada.
