```python
import time
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from pathlib import Path
plt.style.use('default')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['savefig.edgecolor'] = 'white'

plt.rcParams['figure.dpi'] = 500
plt.rcParams['savefig.dpi'] = 500
```

```python

def multiplicacion_matrices(A, B):
    n = len(A)

    C = np.zeros((n, n))

    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]

    return C

def medir_tiempo_promedio_multiplicacion(n, repeticiones=10):
    tiempos = np.zeros(repeticiones)
    for i in range(repeticiones):
        A = np.random.randint(0, 100, size=(n, n))
        B = np.random.randint(0, 100, size=(n, n))
        inicio = time.perf_counter()
        multiplicacion_matrices(A, B)
        fin = time.perf_counter()
        tiempos[i] = fin - inicio
    return np.average(tiempos)
```

```python

listas = np.arange(1, 105, 1)

tiempos_promedio_multiplicacion = np.zeros(len(listas))

for i, n in enumerate(listas):
    tiempos_promedio_multiplicacion[i] = medir_tiempo_promedio_multiplicacion(n)

cubica_listas = listas ** 3
slope_cubica, intercept_cubica, r_value_cubica, p_value_cubica, std_err_cubica = stats.linregress(cubica_listas, tiempos_promedio_multiplicacion)

cubica_ajustada_tiempos = slope_cubica * cubica_listas + intercept_cubica

fig, ax1 = plt.subplots(1, 1, figsize=(8, 4))

ax1.plot(listas, tiempos_promedio_multiplicacion, label='Función de complejidad experimental')

ax1.plot(listas, cubica_ajustada_tiempos, label=f'Función de complejidad teórica $T(n) = n^3$', linestyle='dotted', color='red')

ax1.set_xlabel('Tamaño de la entrada ($n$)')
ax1.set_ylabel('Tiempo de ejecución promedio [s]')
ax1.set_xlim(left=0)
ax1.set_ylim(bottom=0)
ax1.set_title(f'T(n) teórico vs T(n) calculado')
ax1.legend(loc='upper left')
ax1.grid(True)

ax1.xaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
ax1.yaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
ax1.ticklabel_format(axis='x', style='sci', scilimits=(0, 0))
ax1.ticklabel_format(axis='y', style='sci', scilimits=(0, 0))

plt.tight_layout()
plt.show()

Path("graficas").mkdir(parents=True, exist_ok=True)
fig.savefig('graficas/complejidad_cubica.png', bbox_inches="tight", pad_inches=.05)
```
