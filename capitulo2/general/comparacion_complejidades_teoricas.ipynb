```python
import math
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np

plt.style.use('default')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['savefig.edgecolor'] = 'white'
plt.rcParams['figure.dpi'] = 500
plt.rcParams['savefig.dpi'] = 500

Path('graficas').mkdir(exist_ok=True)

FIGSIZE = (10, 6)
FIG_DPI = 500
```

```python
n = np.arange(1, 501)
limite_superior = 10**10

complejidades = [
    (r'$1$', np.ones_like(n, dtype=float)),
    (r'$\log_{\ell}(n)$', np.log2(n)),
    (r'$n$', n.astype(float)),
    (r'$n \log_{\ell}(n)$', n * np.log2(n)),
    (r'$n^2$', n.astype(float) ** 2),
    (r'$n^3$', n.astype(float) ** 3),
    (r'$2^n$', 2.0 ** np.minimum(n, 40)),
    (r'$n!$', np.array([math.factorial(int(valor)) if valor <= 20 else np.nan for valor in n], dtype=float)),
]

posiciones_etiquetas = {
    r'$1$': 1.45,
    r'$\log_{\ell}(n)$': np.log2(500),
    r'$n$': 500,
    r'$n \log_{\ell}(n)$': 500 * np.log2(500),
    r'$n^2$': 500**2,
    r'$n^3$': 500**3,
    r'$2^n$': 10**9,
    r'$n!$': 3.8 * 10**9,
}


def configurar_ejes(ax):
    ax.set_title('Comparación de complejidades teóricas', fontsize=13, pad=8)
    ax.set_xlabel('Tamaño de la entrada ($n$)', fontsize=11, labelpad=6)
    ax.set_ylabel('Número de operaciones', fontsize=11)
    ax.set_yscale('log')
    ax.set_xlim(1, 500)
    ax.set_ylim(1, limite_superior)
    ax.set_xticks([100, 200, 300, 400, 500])
    ax.set_yticks([10**0, 10**2, 10**4, 10**6, 10**8, 10**10])
    ax.grid(True, which='major', linestyle='--', linewidth=0.6, color='#b8b8b8', alpha=0.7)
    ax.tick_params(axis='both', labelsize=9, colors='#111111')


def agregar_etiquetas(ax):
    for etiqueta, y_pos in posiciones_etiquetas.items():
        ax.text(505, y_pos, etiqueta, va='center', ha='left', fontsize=10, clip_on=False)


def guardar_figura(fig, nombre_archivo):
    fig.set_size_inches(*FIGSIZE, forward=True)
Path("graficas").mkdir(parents=True, exist_ok=True)
    fig.savefig(f'graficas/{nombre_archivo}', dpi=FIG_DPI, bbox_inches='tight', pad_inches=0)
```

```python
fig, ax = plt.subplots(figsize=FIGSIZE, dpi=FIG_DPI)

for etiqueta, valores in complejidades:
    valores_graficados = np.where(valores > limite_superior, np.nan, valores)
    ax.plot(n, valores_graficados, color='#222222', linewidth=1.2)

configurar_ejes(ax)

for spine in ax.spines.values():
    spine.set_color('#777777')
    spine.set_linewidth(0.9)

agregar_etiquetas(ax)
fig.subplots_adjust(left=0.08, right=0.86, bottom=0.14, top=0.90)
guardar_figura(fig, 'complejidad_teorica.png')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=FIGSIZE, dpi=FIG_DPI)

zonas = [
    (1, posiciones_etiquetas[r'$\log_{\ell}(n)$'], '#55ff5f', 'Excelente'),
    (posiciones_etiquetas[r'$\log_{\ell}(n)$'], posiciones_etiquetas[r'$n$'], '#fbff4c', 'Bueno'),
    (posiciones_etiquetas[r'$n$'], posiciones_etiquetas[r'$n \log_{\ell}(n)$'], '#ffe45c', 'Aceptable'),
    (posiciones_etiquetas[r'$n \log_{\ell}(n)$'], posiciones_etiquetas[r'$n^2$'], '#ffc85c', 'Malo'),
    (posiciones_etiquetas[r'$n^2$'], limite_superior, '#ff5d61', 'Pésimo'),
]

for y_min, y_max, color, etiqueta in zonas:
    ax.axhspan(y_min, y_max, facecolor=color, edgecolor=color, alpha=0.95, label=etiqueta)

for etiqueta, valores in complejidades:
    valores_graficados = np.where(valores > limite_superior, np.nan, valores)
    ax.plot(n, valores_graficados, color='black', linewidth=1.2)

configurar_ejes(ax)

for spine in ax.spines.values():
    spine.set_color('#333333')
    spine.set_linewidth(0.9)

agregar_etiquetas(ax)
ax.legend(
    loc='upper center',
    bbox_to_anchor=(0.5, 0.985),
    ncol=5,
    frameon=True,
    framealpha=1,
    edgecolor='#999999',
    fontsize=9,
    borderpad=0.35,
    labelspacing=0.25,
    handletextpad=0.45,
    columnspacing=0.9,
)

fig.subplots_adjust(left=0.08, right=0.86, bottom=0.14, top=0.90)
guardar_figura(fig, 'complejidad_ejecucion.png')
plt.show()
```
