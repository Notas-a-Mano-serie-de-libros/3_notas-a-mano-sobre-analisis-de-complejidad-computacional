```python
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
plt.style.use('default')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['savefig.edgecolor'] = 'white'

plt.rcParams['figure.dpi'] = 500
plt.rcParams['savefig.dpi'] = 500
```

```python

k_max_1 = 4
ylim_1 = 4
valores_k_1 = np.arange(0, k_max_1 + 1, 1)
tamanos_problema_1 = np.linspace(1, 10, 10**ylim_1)

fig1, ax1 = plt.subplots(1, 1, figsize=(8, 4))

ax1.set_ylim([0, 10])
ax1.set_xlim([1, 10.6])

annotate_positions_1 = {
    4: (1.2, 8),
    3: (2.0, 8),
    2: (2.9, 8),
    1: (8.4, 8),
    0: (10, 1)
}

for k in valores_k_1:
    complejidad_teorica = tamanos_problema_1 ** k
    line, = ax1.plot(tamanos_problema_1, complejidad_teorica, label=f'$n^{{{k}}}$')

    if k in annotate_positions_1:
        ax1.annotate(f'$n^{{{k}}}$', xy=annotate_positions_1[k],
                     xytext=(5, 0), textcoords='offset points', fontsize=14, color=line.get_color())

ax1.set_xlabel('Tamaño de la entrada ($n$)')
ax1.set_ylabel('Función de complejidad teórica')
ax1.set_title(f'$C(n)=n^k$ para $k \\in [0, {ylim_1}]$')
ax1.grid(True)

ax1.xaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
ax1.yaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))

plt.tight_layout()
plt.show()

Path("graficas").mkdir(parents=True, exist_ok=True)
fig1.savefig('graficas/complejidad_teorica_polinomica1.png', bbox_inches="tight", pad_inches=.05)
```

```python

k_max_2 = 10
ylim_2 = 6
valores_k_2 = np.arange(0, k_max_2 + 1, 1)
tamanos_problema_2 = np.linspace(1, 10, 10**ylim_2)

fig2, ax2 = plt.subplots(1, 1, figsize=(8, 4))

ax2.set_ylim([-10**5, 10**ylim_2])
ax2.set_xlim([2, 10.8])

valor_referencia_2 = 0.5 * 10**ylim_2

etiqueta_n_0_k_añadida = False

for k in valores_k_2:
    complejidad_teorica = tamanos_problema_2 ** k
    line, = ax2.plot(tamanos_problema_2, complejidad_teorica, label=f'$n^{{{k}}}$')

    if k >= ylim_2:
        position_index = np.abs(complejidad_teorica - valor_referencia_2).argmin()
        ax2.annotate(f'$n^{{{k}}}$', xy=(tamanos_problema_2[position_index]+.05, complejidad_teorica[position_index]),
                     xytext=(5, 0), textcoords='offset points', fontsize=14, color=line.get_color())
    elif k == ylim_2 - 1:
        position_index = -1
        ax2.annotate(f'$n^{ylim_2-1}$', xy=(tamanos_problema_2[position_index], complejidad_teorica[position_index]),
                     xytext=(5, 0), textcoords='offset points', fontsize=14, color=line.get_color())
    elif not etiqueta_n_0_k_añadida:
        position_index = -1
        ax2.annotate(f'$n^{{{[0, ylim_2-2]}}}$', xy=(tamanos_problema_2[position_index], complejidad_teorica[position_index]),
                     xytext=(5, 0), textcoords='offset points', fontsize=14, color='black')
        etiqueta_n_0_k_añadida = True

ax2.set_xlabel('Tamaño de la entrada ($n$)')
ax2.set_ylabel('Función de complejidad teórica')
ax2.set_title(f'$C(n)=n^k$ para $k \\in [0, {k_max_2}]$')
ax2.grid(True)

ax2.xaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
ax2.yaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
ax2.ticklabel_format(axis='y', style='sci', scilimits=(0, 0))

plt.tight_layout()
plt.show()

Path("graficas").mkdir(parents=True, exist_ok=True)
fig2.savefig('graficas/complejidad_teorica_polinomica2.png', bbox_inches="tight", pad_inches=.05)
```
