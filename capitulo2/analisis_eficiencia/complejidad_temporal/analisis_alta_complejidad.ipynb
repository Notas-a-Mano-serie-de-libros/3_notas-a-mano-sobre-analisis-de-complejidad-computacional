```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from pathlib import Path
plt.style.use('default')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['savefig.edgecolor'] = 'white'

plt.rcParams['figure.dpi'] = 500
plt.rcParams['savefig.dpi'] = 500
plt.subplots_adjust(left=0, right=1, top=1, bottom=0)

n_values = np.logspace(0, 4, num=5, dtype=int)
```

```python

fig, ax2 = plt.subplots(1, 1, figsize=(8, 5))

x = [1e0, 1e1, 1e2, 1e3, 1e4]
y = [0.5, 0.8, 24, 295, 3004]

log_x = np.log10(x)
log_y = np.log10(y)
interp_func_2 = interp1d(log_x, log_y, fill_value="extrapolate")
log_x_for_y_17_2 = interp_func_2(np.log10(17))
x_for_y_17_2 = 10 ** log_x_for_y_17_2

ax2.set_yscale('log')
ax2.set_xscale('log')
ax2.set_ylim(1e-1, 1e4)

yticks = [1E-1, 1e0, 1e1, 1e2, 1e3, 1e4]
ytick_labels = [r'$10^{-10}$', r'$10^{0}$', r'$10^{10}$', r'$10^{100}$', r'$10^{1000}$', r'$10^{10000}$']
plt.yticks(yticks, ytick_labels)

ax2.set_xlabel('Tamaño de la entrada (n)')
ax2.set_ylabel('Tiempo de ejecución teórico [s]')
ax2.set_title('Comportamiento de la función $T(n) = 2^n$')
ax2.plot(x, y, marker='o', linestyle='-', color='b', label = 'Complejidad temporal teórica')

ax2.scatter(76, 17, color='red', marker='*', s=100, label='Edad estimada del Universo: $4.17 \\cdot 10^{17}$ [s]', zorder=5)
ax2.grid(True, which="both", ls="--")

ax2.legend(loc='upper left')

plt.show()
Path("graficas").mkdir(parents=True, exist_ok=True)
fig.savefig('graficas/complejidad_exponencial_1.png', bbox_inches="tight", pad_inches=.05)
```

```python

fig, ax2 = plt.subplots(1, 1, figsize=(8, 3))

x_new = [76, 100]
y_new = [10**17, 10**24]

ax2.set_yscale('log')

ax2.set_ylim(1e16, 1e25)
ax2.set_xlim(75, 101)

ax2.set_xlabel('Tamaño de la entrada (n)')
ax2.set_ylabel('Tiempo de ejecución teórico [s]')
ax2.set_title('Comportamiento de la función $T(n)$ para valores grandes')

ax2.scatter(x_new[0], y_new[0], color='red', marker='*', s=150, label='Edad estimada del Universo: $4.17 \\cdot 10^{17}$ [s]', zorder=5)

ax2.scatter(x_new[1], y_new[1], color='blue', marker='o', s=100, label='Tiempo de ejecución para $n=100$')

ax2.grid(True, which="both", ls="--")

ax2.legend(loc='upper left')

plt.show()

Path("graficas").mkdir(parents=True, exist_ok=True)
fig.savefig('graficas/complejidad_exponencial_2.png', bbox_inches="tight", pad_inches=.05)
```

```python

fig, ax2 = plt.subplots(1, 1, figsize=(8, 5))

x = np.array([1e0, 1e1, 1e2, 1e3, 1e4])
y = np.array([0.5, 1, 151, 2561, 35653])

interpolation_function = interp1d(y, x, kind='linear', fill_value="extrapolate")

y_target = 17
x_for_y_17 = interpolation_function(y_target)

ax2.set_yscale('log')
ax2.set_xscale('log')
ax2.set_ylim(1e-1, 1e5)

yticks = [1E-1, 1e0, 1e1, 1e2, 1e3, 1e4, 1e5]
ytick_labels = [r'$10^{-10}$', r'$10^{0}$', r'$10^{10}$', r'$10^{100}$', r'$10^{1000}$', r'$10^{10000}$', r'$10^{100000}$']
plt.yticks(yticks, ytick_labels)

ax2.set_xlabel('Tamaño de la entrada (n)')
ax2.set_ylabel('Tiempo de ejecución teórico [s]')
ax2.set_title('Comportamiento de la función $T(n) = n!$')
ax2.plot(x, y, marker='o', linestyle='-', color='b', label='Complejidad temporal teórica')

ax2.scatter(36, y_target, color='red', marker='*', s=100, label='Edad estimada del Universo: $4.17 \\cdot 10^{17}$ [s]', zorder=5)
ax2.grid(True, which="both", ls="--")

ax2.legend(loc='upper left')

plt.show()
Path("graficas").mkdir(parents=True, exist_ok=True)
fig.savefig('graficas/complejidad_factorial_1.png', bbox_inches="tight", pad_inches=.05)
```
