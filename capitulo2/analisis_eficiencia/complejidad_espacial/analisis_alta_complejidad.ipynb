```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
plt.style.use('default')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['savefig.edgecolor'] = 'white'

plt.rcParams['figure.dpi'] = 500
plt.rcParams['savefig.dpi'] = 500
plt.subplots_adjust(left=0, right=1, top=1, bottom=0)

n_values = np.logspace(0, 4, num=5, dtype=int)
```

```python

fig, ax2 = plt.subplots(1, 1, figsize=(8, 5))

x = [1e0, 1e1, 1e2, 1e3, 1e4]
y = [1, 10, 30, 301, 3010]

ax2.set_yscale('log')
ax2.set_xscale('log')
ax2.set_ylim(1e-1, 1e4)

yticks = [1E-1, 1e0, 1e1, 1e2, 1e3, 1e4]
ytick_labels = [r'$10^{-10}$', r'$10^{0}$', r'$10^{10}$', r'$10^{100}$', r'$10^{1000}$', r'$10^{10000}$']
plt.yticks(yticks, ytick_labels)

ax2.set_xlabel('Tamaño de la entrada (n)')
ax2.set_ylabel('Consumo de recursos teórico [bytes]')
ax2.set_title('Comportamiento de la función $S(n) = 2^n$')
ax2.plot(x, y, marker='o', linestyle='-', color='b', label = 'Complejidad espacial teórica')

ax2.scatter(50, 21, color='red', marker='x', s=100, label='Cantidad de recursos consumidos por la humanidad: $10^{21}$ [bytes]', zorder=5)
ax2.grid(True, which="both", ls="--")

ax2.legend(loc='upper left')

plt.show()
Path("graficas").mkdir(parents=True, exist_ok=True)
fig.savefig('graficas/complejidad_exponencial_1.png', bbox_inches="tight", pad_inches=.05)
```

```python

fig, ax2 = plt.subplots(1, 1, figsize=(8, 3))

x_new = [51, 100]
y_new = [10**21, 10**30]

ax2.set_yscale('log')

ax2.set_ylim(1e20, 1e31)
ax2.set_xlim(49, 102)

ax2.set_xlabel('Tamaño de la entrada (n)')
ax2.set_ylabel('Consumo de recursos teórico [bytes]')
ax2.set_title('Comportamiento de la función $S(n)$ para valores grandes')

ax2.scatter(x_new[0], y_new[0], color='red', marker='x', s=150, label='Cantidad de recursos consumidos por la humanidad: $10^{21}$ [bytes]', zorder=5)

ax2.scatter(x_new[1], y_new[1], color='blue', marker='o', s=100, label='Consumo de recursos para $n=100$')

ax2.grid(True, which="both", ls="--")

ax2.legend(loc='upper left')

plt.show()

Path("graficas").mkdir(parents=True, exist_ok=True)
fig.savefig('graficas/complejidad_exponencial_2.png', bbox_inches="tight", pad_inches=.05)
```
