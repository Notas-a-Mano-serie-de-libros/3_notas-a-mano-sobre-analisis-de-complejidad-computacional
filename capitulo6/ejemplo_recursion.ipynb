```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('default')
plt.rcParams['figure.dpi']        = 500
plt.rcParams['savefig.dpi']       = 500
plt.rcParams['figure.facecolor']  = 'white'
plt.rcParams['axes.facecolor']    = 'white'
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['savefig.edgecolor'] = 'white'

def aplicar_funcion(n, funcion):
    valores = [n]
    while n > 1:
        n = funcion(n)
        valores.append(n)
    return valores

n_inicial = 100

valores_n1 = aplicar_funcion(n_inicial, lambda x: x - 1)
valores_n2 = aplicar_funcion(n_inicial, lambda x: x - 2)

pasos_n1 = len(valores_n1) - 1
pasos_n2 = len(valores_n2) - 1

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(valores_n1, color='#2563eb', linewidth=2,
        label=rf'$C(n-1)$ — {pasos_n1} pasos')
ax.plot(valores_n2, color='#dc2626', linewidth=2,
        linestyle='-', label=rf'$C(n-2)$ — {pasos_n2} pasos')

ax.set_xlabel('Pasos', fontsize=13)
ax.set_ylabel('Valor de $n$', fontsize=13)
ax.set_title(rf'Decrecimiento desde $n = {n_inicial}$ aplicando $C(n-1)$ y $C(n-2)$',
             fontsize=14)
ax.legend(fontsize=11)
ax.grid(True, linestyle='-', alpha=0.4)

plt.tight_layout()

fig.savefig('graficas/comparacion_llamados_recursivos.png',
            bbox_inches='tight', pad_inches=0.05)

plt.show()
```
