```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('default')
plt.rcParams['figure.dpi'] = 500
plt.rcParams['savefig.dpi'] = 500
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'
plt.rcParams['savefig.edgecolor'] = 'white'


n = np.linspace(1, 30, 500)

phi = (1 + np.sqrt(5)) / 2
exact    = phi**n                    # Θ(φⁿ) — solución exacta
upper    = 2**n                      # O(2ⁿ)  — cota superior
lower    = (np.sqrt(2))**n           # Ω(√2ⁿ) — cota inferior

fig, ax = plt.subplots(1, 1, figsize=(8, 4))

ax.semilogy(n, exact, color='#2563eb', linewidth=2.5,
            label=r'$\Theta(\varphi^n)$ — solución exacta')
ax.semilogy(n, upper, color='#dc2626', linewidth=2,
            linestyle='--', label=r'$O(2^n)$ — cota superior')
ax.semilogy(n, lower, color='#16a34a', linewidth=2,
            linestyle='--', label=r'$\Omega((\sqrt{2})^n)$ — cota inferior')

ax.fill_between(n, lower, upper, alpha=0.07, color='#2563eb')

ax.set_xlabel('$n$', fontsize=13)
ax.set_ylabel('$T(n)$', fontsize=13)
ax.set_title('Fibonacci recursivo — cotas asintóticas', fontsize=14)
ax.legend(fontsize=11)
ax.grid(True, which='both', linestyle='--', alpha=0.4)

plt.tight_layout()

fig.savefig('graficas/comparacion_fibonacci.png', bbox_inches="tight", pad_inches=.05)

plt.show()
```
